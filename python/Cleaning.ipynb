```python
import pandas as pd

df = pd.read_csv("../Data/Super store data set.csv")
```

```python
df = df.drop_duplicates()
```

```python
df.shape
```

```python
df["Order Date"] = pd.to_datetime(df["Order Date"], dayfirst=True)
df["Ship Date"] = pd.to_datetime(df["Ship Date"], dayfirst=True)
```

```python
df.info()
```

```python
df["Year"] = df["Order Date"].dt.year
df["Month"] = df["Order Date"].dt.month_name()
df["Month No"] = df["Order Date"].dt.month
df["Quarter"] = df["Order Date"].dt.quarter
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.to_csv("../Data/Cleaned_Superstore.csv", index=False)
```

```python
total_sales = df["Sales"].sum()
print("Total Sales:",total_sales)
```

```python
total_orders = df["Order ID"].nunique()
print("Total Orders:", total_orders)
```

```python
total_customers = df["Customer ID"].nunique()
print("Total Customers:", total_customers)
```

```python
sales_by_category = df.groupby("Category")["Sales"].sum().sort_values(ascending=False)
print(sales_by_category)
```

```python
sales_by_region = df.groupby("Region")["Sales"].sum().sort_values(ascending=False)
print(sales_by_region)
```

```python
top_products = (
    df.groupby("Product Name")["Sales"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

print(top_products)
```

```python
sales_by_segment = df.groupby("Segment")["Sales"].sum()
print(sales_by_segment)
```

```python
sales_by_state = df.groupby("State")["Sales"].sum().sort_values(ascending=False)
print(sales_by_state.head(10))
```
